The expressions for the vectors as given in the project are:

$$\label{eq:1}
\displaystyle\vec{r}_1 =
\begin{bmatrix}
    \displaystyle l_1\sin{(\theta)} \\
    \displaystyle -l_1\cos{(\theta)} + l_3
\end{bmatrix}
$$

$$\label{eq:2}
\displaystyle\vec{r}_{2a} =
\begin{bmatrix}
    \displaystyle -l_2\sin{(\theta)} \\
    \displaystyle l_2\cos{(\theta)} + l_3
\end{bmatrix}
$$

$$\label{eq:3}
\displaystyle\vec{r}_{2b} =
\begin{bmatrix}
    \displaystyle -l_2\sin{(\theta)} - l_4\sin{(\psi - \theta)} \\
    \displaystyle l_2\cos{(\theta)} + l_3 - l_4\cos{(\psi - \theta)}
\end{bmatrix}
$$

To calculate we need $\displaystyle T_i = \frac12m_iv_i$, where $\displaystyle v_i = \frac{d\vec{r}_i}{dt}.$

$$
\displaystyle\vec{v}_1 = \frac{d\vec{r}_1}{dt} = 
\begin{bmatrix}
    \displaystyle l_1\theta\cos{(\theta)} \\
    \displaystyle l_1\theta\sin{(\theta)}
\end{bmatrix},
$$

which gives,

$$\label{eq:4}
    \displaystyle v_1^2 = (l_1\theta\cos{(\theta)})^2 + (l_1\theta\sin{(\theta)})^2 = (l_1\theta)^2.
$$

$$
\displaystyle\vec{v}_{2a} = \frac{d\vec{r}_{2a}}{dt} = 
\begin{bmatrix}
    \displaystyle -l_2\theta\cos{(\theta)} \\
    \displaystyle -l_2\theta\sin{(\theta)}
\end{bmatrix},
$$

which gives,

$$\label{eq:5}
    \displaystyle v_{2a}^2 = (-l_2\theta\cos{(\theta)})^2 + (-l_2\theta\sin{(\theta)})^2 = (l_2\theta)^2.
$$

Since $\vec{r}_{2a}$ is a point, and considered massless. It has no contribution to the kinetic energy.

$$
\displaystyle\vec{v}_{2b} = \frac{d\vec{r}_{2b}}{dt} = 
\begin{bmatrix}
    \displaystyle -l_2\theta\cos{(\theta)} - l_4(\psi - \theta)\cos{(\psi - \theta)} \\
    \displaystyle -l_2\theta\sin{(\theta)} + l_4(\psi - \theta)\sin{(\psi - \theta)}
\end{bmatrix},
$$

which gives,

$$\label{eq:6}
\begin{matrix}
    \displaystyle v_{2b}^2 = (-l_2\theta\cos{(\theta)} - l_4(\psi - \theta)\cos{(\psi - \theta)})^2 + (-l_2\theta\sin{(\theta)} + l_4(\psi - \theta)\sin{(\psi - \theta)})^2 \\ 
    = (l_2\theta)^2 + (l_4(\psi - \theta))^2 + \left[2l_2\theta\cos{(\theta)}l_4(\psi - \theta)\cos{(\psi - \theta)}\right] - \left[2l_2\theta\sin{(\theta)} l_4(\psi - \theta)\sin{(\psi - \theta)}\right],
\end{matrix}
$$

Using $\sin{(\psi - \theta)} = \sin{(\psi)}\cos{(\theta)} - \cos{(\psi)}\sin{(\theta)}$ and $\cos{(\psi - \theta)} = \cos{(\psi)}\cos{(\theta)} + \sin{(\psi)}\sin{(\theta)}$, we get:

$$\label{eq:7}
    \displaystyle v_{2b}^2 = (l_2\theta)^2 + (l_4(\psi - \theta))^2 + 2l_2l_4\theta(\psi - \theta)\cos{(\psi)}.
$$

Combining the equation for $v_1$ and $v_{2b}$ the kinetic energy becomes:

$$\label{eq:8}
    \displaystyle T = \frac12\left(m_1l_1^2 + m_2l_2^2\right)\theta^2 + \frac12m_2(l_4(\psi - \theta))^2 + m_2l_2l_4\theta(\psi - \theta)\cos{(\psi)}.
$$

To find the potential we need to consider the two masses in our system. Since we know $g$, $m_1$ and $m_2$, all we need to do, is to find their heights as a function of $\theta$ and $\psi$. Since $h_1 = l_3 - l_1\cos{(\theta)}$, we get:

$$\label{eq:9}
   \displaystyle V_1 = gm_1h_1 = gm_1\left( l_3 - l_1\cos{(\theta)} \right).
$$

Since $h_2 = \left( l_3 + l_2\cos{(\theta)} - l_4\cos{(\psi - \theta)}\right)$, we get:

$$\label{eq:10}
    \displaystyle V_2 = gm_2h_2 = gm_2\left( l_3 + l_2\cos{(\theta)} - l_4\cos{(\psi - \theta)}\right).
$$

The total potential then becomes:

$$\label{eq:11}
    \displaystyle V = V_1 + V_2 = gl_3\left( m_1 + m_2 \right) - g\left( m_1l_1 - m_2l_2 \right)\cos{(\theta)} - gm_2l_4\cos{(\psi-\theta)}.
$$

By combining the equations for kinetic and potential energy the Lagrangian becomes:

$$\label{eq:12}
\begin{matrix}    
    \displaystyle L\left(\theta, \theta, \psi, \psi \right) = T - V \\
    = \frac12\left(m_1l_1^2 + m_2l_2^2\right)\theta^2 + \frac12m_2(l_4(\psi - \theta))^2 + m_2l_2l_4\theta(\psi - \theta)\cos{(\psi)} \\
    - gl_3\left( m_1 + m_2 \right) + g\left( m_1l_1 - m_2l_2 \right)\cos{(\theta)} + gm_2l_4\cos{(\psi-\theta)}.
\end{matrix}
$$

To find the equations of motion, we use the Euler-Lagrange equations, with generalized coordinates $q = [\theta, \psi]$.

$$\label{eq:13}
    \displaystyle \frac{\partial L}{\partial \theta} = - g\left( m_1l_1 - m_2l_2 \right)\sin{(\theta)} + gm_2l_4\sin{(\psi-\theta)},
$$

$$\label{eq:14}
\begin{matrix}
    \displaystyle \frac{\partial}{\partial t}\left(\frac{\partial L}{\partial \theta}\right) = \frac{\partial}{\partial t}\left( \left(m_1l_1^2 + m_2l_2^2\right)\theta - m_2l_4^2(\psi - \theta) + m_2l_2l_4\cos{(\psi)}(\psi - 2\theta) \right) \\
    = \left(m_1l_1^2 + m_2l_2^2\right)\ddot{\theta} - m_2l_4^2(\ddot{\psi} - \ddot{\theta}) + m_2l_2l_4(-\psi(\psi - 2\theta)\sin{(\psi)} + (\ddot{\psi} - 2\ddot{\theta})\cos{(\psi)}) \\
    = \ddot{\theta}\left( m_1l_1^2 + m_2l_2^2 + m_2l_4^2 - 2m_2l_2l_4\cos{(\psi)} \right) + \ddot{\psi}\left( m_2l_2l_4\cos{(\psi)} - m_2l_4^2 \right) \\
    \quad - \psi(\psi - 2\theta)m_2l_2l_4\sin{(\psi)},
\end{matrix}
$$

$$\label{eq:15}
\begin{matrix}
    \displaystyle \frac{\partial L}{\partial \psi} = - m_2l_2l_4\theta(\psi - \theta)\sin{(\psi)} - gm_2l_4\sin{(\psi - \theta)},
\end{matrix}
$$

$$\label{eq:16}
\begin{matrix}
    \displaystyle \frac{\partial}{\partial t}\left(\frac{\partial L}{\partial \psi}\right) = \frac{\partial}{\partial t}\left( m_2l_4^2(\psi - \theta) + m_2l_2l_4\theta\cos{(\psi)} \right) \\
    = m_2l_4^2(\ddot{\psi} - \ddot{\theta}) + m_2l_2l_4(\ddot{\theta}\cos{(\psi)} - \psi\theta\sin{(\psi)}) \\
    = \ddot{\theta}m_2l_4(l_2\cos{(\psi)} - l_4) + \ddot{\psi}m_2l_4^2 - \psi\theta m_2l_2l_4\sin{(\psi)}.
\end{matrix}
$$

The equations of motion becomes:

$$\label{eq:17}
\begin{matrix}
    - g\left( m_1l_1 - m_2l_2 \right)\sin{(\theta)} + gm_2l_4\sin{(\psi-\theta)} 
    - [ \ddot{\theta}\left( m_1l_1^2 + m_2l_2^2 + m_2l_4^2 - 2m_2l_2l_4\cos{(\psi)} \right) \\ + \ddot{\psi}\left( m_2l_2l_4\cos{(\psi)} - m_2l_4^2 \right)
    - \psi(\psi - 2\theta)m_2l_2l_4\sin{(\psi)} ] = 0,
\end{matrix}
$$

and

$$\label{eq:18}
\begin{matrix}
    - m_2l_2l_4\theta(\psi - \theta)\sin{(\psi)} - gm_2l_4\sin{(\psi - \theta)} \\
    - \left( \ddot{\theta}m_2l_4(l_2\cos{(\psi)} - l_4) + \ddot{\psi}m_2l_4^2 - \psi\theta m_2l_2l_4\sin{(\psi)} \right) = 0.
\end{matrix}
$$

Which can be rearranged to:

$$\label{eq:19}
\begin{matrix}
    \quad \ddot{\theta}\left( m_1l_1^2 + m_2l_2^2 + m_2l_4^2 - 2m_2l_2l_4\cos{(\psi)} \right) + \ddot{\psi}\left( m_2l_2l_4\cos{(\psi)} - m_2l_4^2 \right) \\
    = - g\left( m_1l_1 - m_2l_2 \right)\sin{(\theta)} + gm_2l_4\sin{(\psi-\theta)} + \psi(\psi - 2\theta)m_2l_2l_4\sin{(\psi)} \\
    = - gm_1l_1\sin{(\theta)} + gm_2(l_2\sin{(\theta)} + l_4\sin{(\psi - \theta)}) + m_2l_2l_4\psi(\psi - 2\theta)\sin{(\psi)},
\end{matrix}
$$
 
and

$$\label{eq:20}
\begin{matrix}
    \ddot{\theta}m_2l_4(l_2\cos{(\psi)} - l_4) + \ddot{\psi}m_2l_4^2 = m_2l_2l_4\theta^2\sin{(\psi)} - gm_2l_4\sin{(\psi - \theta)} \\
    = m_2l_4(l_2\theta^2\sin{(\psi)} - g\sin{(\psi - \theta)}).
\end{matrix}
$$

We can then express the equations of motion as an equation of matrices like below:

$$\label{eq:21}
\begin{bmatrix}
    a_{11} a_{12} \\
    a_{21} a_{22}
\end{bmatrix}
\begin{bmatrix}
    \ddot{\theta} \\
    \ddot{\psi}
\end{bmatrix} = 
\begin{bmatrix}
    b_1 \\
    b_2
\end{bmatrix}.
$$

Where:

$$\label{eq:22}
    \begin{matrix}
        a_{11}(\psi) = m_1l_1^2 + m_2l_2^2 + m_2l_4^2 - 2m_2l_2l_4\cos{(\psi)}, \\
        a_{12}(\psi) = m_2l_4(l_2\cos{(\psi)} - l_4), \\
        a_{21}(\psi) = m_2l_4(l_2\cos{(\psi)} - l_4) = a_{12}(\psi), \\
        a_{22} = m_2l_4^2.
    \end{matrix}
$$

```
## Constants for later use

m_1 = 2000 # kg, mass of the counterweight
m_2 = 15 # kg, mass of the projectile
l_1 = 1.2 # m, short part of the main beam
l_2 = 5.7 # m, long part of the main beam
l_3 = 3.2 # m, stand
l_4 = 5 # m, projectile sling

g = 9.81 # m/s^2 gravitational constant

theta_0 = 0.7 * np.pi # rad
theta_0_dot = 0 # rad/s
theta_r = 0.1 * np.pi  # Release angle 

psi_0 = theta_0 - 0.5*np.pi # rad
psi_0_dot = 0 # rad/s
```

```
def ode_solver(f, q0, t0, T, mode, theta_r = 0, step_length=1e-3):
    '''
    parameters: f, the system we eould like to solve.
                q0, initial conditions.
                t0, end initial time.
                step_length, length of each iteration step.
    returns: the solution of the system for every t.
    '''

    #Initializing:
    q_num = np.array([q0])    # Array for q-values
    t_num = np.array([t0])    # Array for time

    qn = q0                # temporary values for q and t
    tn = t0

    if mode == "theta_r":
    
        # Main loop
        while qn[0] > theta_r:    # Release angle     
            qn, tn = RK4(f, qn, tn, step_length)    # One step of Runge-Kutta 4 method
            q_num = np.concatenate((q_num, np.array([qn])))
            t_num = np.append(t_num, tn)
    
    elif mode == "projectile motion":

        while qn[2] > 0:    # Until projectile hits ground     
            qn, tn = RK4(f, qn, tn, step_length)    # One step of Runge-Kutta 4 method
            q_num = np.concatenate((q_num, np.array([qn])))
            t_num = np.append(t_num, tn)
    else:
        while tn < T:
            qn, tn = RK4(f, qn, tn, step_length)    # One step of Runge-Kutta 4 method
            q_num = np.concatenate((q_num, np.array([qn])))
            t_num = np.append(t_num, tn) 



    return q_num.T, t_num

# Runga Kutta of 4th order
def RK4(f, q, t, step_length=1e-2):
    """
    parameters: f, the system we would like to solve.
                q, postion vector, q = [x,y,z]
                t, general variable
                step_length, length of each iteration step.
    returns: next time step with corresponding solution.
    """
    k1 = f(t, q)
    k2 = f(t + 0.5*step_length, q + 0.5*step_length*k1)
    k3 = f(t + 0.5*step_length, q + 0.5*step_length*k2)
    k4 = f(t + step_length, q + step_length*k3)

    q_next = q + step_length*(k1 + 2*k2 + 2*k3 + k4)/6
    t_next = t + step_length
    
    return q_next, t_next
```

```
def A(psi):
    
    a_11 = - l_1**2 * m_1 - l_2**2 * m_2 + 2 * l_2 * l_4 * np.cos(psi) - l_4**2 *m_2
    a_12 = - l_4 * m_2 * (l_2 * np.cos(psi) - l_4)
    a_22 = - l_4**2 * m_2
    
    matrix = np.array([[a_11, a_12], [a_12, a_22]])
    
    return matrix

def b(theta, theta_dot, psi, psi_dot):
    
    b_1 = (g * m_1 * l_1 * np.sin(theta) - g * m_2 * (l_2 * np.sin(theta) + l_4 * np.sin(psi - theta)) 
           - l_2 * l_4 * m_2 *(psi_dot - 2 * theta_dot) * np.sin(psi) * psi_dot )
        
    b_2 = l_4 * m_2 * (g * np.sin(psi - theta) - l_2 * np.sin(psi) * theta_dot**2)
    
    vec = np.array([b_1, b_2])
    
    return vec

def r(theta, psi):
    '''
    parameters: theta, rad
                psi, rad

    returns:  r_arr, (3 x 2) array of correspondingly r_1, r_2a, r_2b
    '''

    r_1 = np.array([l_1 * np.sin(theta), - l_1 * np.cos(theta) + l_3])
    
    r_2a = np.array([-l_2 * np.sin(theta), l_2 * np.cos(theta) + l_3])

    r_2b = r_2a - np.array([l_4 * np.sin(psi - theta), l_4 * np.cos(psi - theta)])


    r_arr = np.array([r_1, r_2a, r_2b])
    return r_arr

def v(theta, theta_dot, psi, psi_dot):

    v_x = - l_2 * np.cos(theta) * theta_dot - l_4 * np.cos(psi - theta) * (psi_dot - theta_dot)
    v_y = - l_2 * np.sin(theta) * theta_dot + l_4 * np.sin(psi - theta) * (psi_dot - theta_dot)

    v = np.array([v_x, v_y])

    return v

def V(m, h):
    return m * g * h


def T(m, v):
    return 1/2 * m * v**2

```

We have the following system:

$$q(t)=[\theta(t) \hspace{0.2cm} \dot{\theta}(t) \hspace{0.2cm} \psi(t)  \hspace{0.2cm} \dot{\psi}(t)],$$

which gives the following derivative:

$$f(t, q) = \frac{dq}{dt} = [\dot{\theta}(t) \hspace{0.2cm} \ddot{\theta}(t) \hspace{0.2cm} \dot{\psi}(t) \hspace{0.2cm} \ddot{\psi}(t)]$$

```
def f(t, q):
    '''
    parameters: t, timestep
                q, np.array(theta_0_dot, theta_0_doubledot, psi_0_dot, psi_0_double_dot)
    '''
    
    d_angle = np.array([q[1], q[3]]) + t * np.linalg.inv(A(q[2])) @ b(q[0], q[1], q[2], q[3])
    dd_angle = np.linalg.inv(A(q[2])) @ b(q[0], q[1], q[2], q[3])

    df = np.array([d_angle[0], dd_angle[0], d_angle[1], dd_angle[1]])
    
    return df
```

The functions are not periodic. They are damped oscillations where $\theta$ goes to $0$ and $\psi$ goes to $2\pi$. That means that the beam will become vertical, and $m_2$ will make an entire rotation before hanging down so that $l_4$ is vertical as well. Since the beam is a damped oscillator, $\dot{\theta}$ and $\dot{\psi}$ will both converge to $0$. So that the beam is at rest, when it is vertical with $m_1$ (the bigger mass) at the bottom.

$$f(t, q) = \frac{dq}{dt} = [v_x(t) \hspace{0.2cm} a_x(t) \hspace{0.2cm} v_y(t) \hspace{0.2cm} a_y(t)]$$

```
def f_pro(t, q):
    
    v = np.array([q[1], q[3]])

    dq = np.array([q[1], 0, q[3], - g])

    return dq
```

```
def abs_v(v_x, v_y):
    v = np.linalg.norm((v_x, v_y))

    return v

def trebuchet(q_0, theta_r_0, theta_r_end, num_steps, mode, title, T_arr = 10):

    theta_r = np.linspace(theta_r_0, theta_r_end, num_steps) * np.pi/ 180

    v_y_min = 10  # Start high

    theta_arr = np.zeros(1)
    v_final_arr = np.zeros_like(theta_r)
    range_arr = np.zeros_like(theta_r)

    if mode != "No plot":
        plt.figure()
        plt.title(title)

    for i, theta in enumerate(theta_r):
        q, t = ode_solver(f, q_0, 0, T, "theta_r", theta)

        r_initial = r(q[0,-1], q[2, -1])[2]
        v_initial = v(q[0,-1], q[1, -1], q[2, -1], q[3, -1])

        q_0_pro = np.array([r_initial[0], v_initial[0], r_initial[1], v_initial[1]])

        q_pro, t_pro = ode_solver(f_pro, q_0_pro, 0, T, "projectile motion")

        v_final = abs_v(q_pro[1], q_pro[3])
        v_final_arr[i] = v_final

        range_arr[i] = q_pro[0, -1]

        if  v_initial[1] < 2 and v_initial[1] > 0: 

            theta_arr = np.append(theta_arr, theta / np.pi * 180)

            if v_initial[1] < v_y_min:
                v_y_min = v_initial[1]
           

        if mode == "xy":

            plt.plot(q_pro[0], q_pro[2], label = f"θ_r = {(theta/np.pi * 180):.2f} deg")
            plt.xlabel("x, m"); plt.ylabel("y,m")

        elif mode == "xt":
            plt.plot(t_pro, q_pro[0], label = f"θ_r = {(theta/np.pi * 180):.2f} deg")
            plt.xlabel("t, s"); plt.ylabel("x,m")
            
        
        elif mode == "yt":
            plt.plot(t_pro, q_pro[2], label = f"θ_r = {(theta/np.pi * 180):.2f} deg")
            plt.xlabel("t, s"); plt.ylabel("y,m")
    
    if mode != "No plot":
        plt.legend()
        plt.show()

    return theta_r, theta_arr, v_final_arr, range_arr, v_y_min
```

```
theta_r, theta_arr, v_arr, _, _ = trebuchet(q_0_2, 0, 40, 10, "xy", "Projectile motion, θ_r ∈ [0, 40] ")
theta_r_2, theta_arr_2, _, _, v_y_min = trebuchet(q_0_2, 0, 1.5, 10, "xy", "Projectile motion θ_r ∈ [0, 1.5]")

print(f'Lowest obtained velocity in y-direction: {v_y_min:.2f} with θ_r ∈ [0, 1.5]')
```

As can be seen in the plots above, the projectile is shot with velocity in only positive $x$-direction when $theta_r$ is low, somewhere in between $(0, 1)^\circ$. Due to the relativetly low number of $\theta_r$ used in this task, there wasn't possible to find a the exact realease angle which gave zero initial velocity of the projectile in $y$-direction. The lowest obtained velocity was $0.95 \text{ m/s}$, but from the trajectory curves with low values of $\theta_r$ that the movement is almost just in $x$-direction from the start.

At impact, the projectile has lost all potential energy, which means that there is only kinetic energy left: $E_{tot} = E_{kin}$.

```
theta_r, _, v_arr, range_arr, _ = trebuchet(q_0_2, 0, 40, 100, "No plot", " ")

E_tot = T(m_2, v_arr)

plt.figure()
plt.title("Energy at impact as function of realease angle")
plt.plot(theta_r, E_tot)
plt.xlabel("θ_r, rad"); plt.ylabel("E, J")
plt.show()

print(f' Highest mechanical energy E_tot = {np.max(E_tot)/1e6:.2f} MJ with release angle {theta_r[np.argmax(E_tot)]:.2f} rad')

plt.figure()
plt.title("Range of projectile in x-direction as function of realease angle")
plt.plot(theta_r, range_arr)
plt.xlabel("θ_r, rad"); plt.ylabel("Range, m")
plt.show()

print(f' Longest range x_max = {np.max(range_arr):.2f} m with release angle {theta_r[np.argmax(range_arr)]:.2f} rad')
```

From the plot of the mechanical energy, $E_{tot}$ at impact as a funtion of release angle $\theta_r$, is at maximal $E_{tot} = 54.76 \text{ MJ}$ when the $\theta_r = 0.22 \text{ rad}$. Likewise the range in $x$-direction is largest, $x_{max} = 173.66 \text{ m}$, when $\theta_r = 0.24$. Since $\theta$ starts high, it makes sence that a large $\theta_r$ corresponds to realeasing the projectile early. Then much of the velocity will work in the vertical plane, and therefore the range will be quite low. However if $\theta_r$ approaches zero, then the air time will be shorter, as there is no movement in $+y$-direction. Therefore the gravity field will do less work on the projectile, implying that the total energy will be lower at impact. That the release angle for the maximum range almost corresponds to the one for maximum mechanical energy, makes sense, as the air time is almost maximal for both caes. The difference is caused by the maximal range needs lower initial velocity in $y$-direction which makes the gravitational field work less on the particle, resulting in a lower kinetic energy.
